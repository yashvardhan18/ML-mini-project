```python
import pandas as pd
```

```python
df=pd.read_csv('tmdb-movies (1).csv')
```

```python
df=df[['id', 'imdb_id', 'popularity', 'budget', 'revenue', 'original_title',
       'cast', 'homepage', 'keywords', 'overview',
       'runtime', 'genres', 'production_companies', 'release_date',
        'release_year', 'budget_adj',
       'revenue_adj']]
```

```python
df.fillna(value=0)
```

```python
df=df.drop(df[(df['budget']<=10000) | (df['revenue']<=10000)].index)
```

```python
df
```

#Q1 Movies with the third highest budget.

```python
df.loc[df['budget']== df.budget.nlargest(3).iloc[-1]]
```

Movies with the third lowest budget.

```python
df.loc[df['budget']== df.budget.nsmallest(3).iloc[-1]]
```

#Q4 Movies with most earned revenue.

```python
df.loc[df['revenue']==df['revenue'].max()]
```

Movies with least earned revenue.

```python
df.loc[df['revenue']==df['revenue'].min()]
```

```python
df.loc[df['release_year']== 2006]
```
